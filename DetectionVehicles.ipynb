```python
img_dir = '/content/VehiclesDetectionDataset/test/images'
img_files = [ f for f in os.listdir(img_dir) if os.path.isfile(os.path.join(img_dir,f))]

for i in range(min(10, len(img_files))):
  img_path = os.path.join(img_dir, img_files[i])
  try:
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(f"Image {i+1}: {img_files[i]}")
    plt.show()

  except Exception as e:
    print(f"Error displaying image {img_files[i]}: {e}")
```
